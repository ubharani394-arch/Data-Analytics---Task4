```python
import pandas as pd
from scipy.stats import ttest_ind

# Load dataset
df = pd.read_excel("ApexPlanet_DataAnalytics_Dataset.xlsx")
# Remove missing values
df = df.dropna()
# Compare Total Sales between Electronics and Fashion
electronics = df[df["Category"] == "Electronics"]["Total_Sales"]
fashion = df[df["Category"] == "Fashion"]["Total_Sales"]
# Perform Independent T-Test
t_stat, p_value = ttest_ind(electronics, fashion)
print("T-Statistic:", t_stat)
print("P-Value:", p_value)
if p_value < 0.05:
    print("Conclusion: There is a statistically significant difference in Total Sales between Electronics and Fashion.")
else:
    print("Conclusion: There is no statistically significant difference in Total Sales between Electronics and Fashion.")
```

```python
import matplotlib.pyplot as plt
df.groupby("Category")["Total_Sales"].sum().plot(kind="pie", autopct="%1.1f%%")
plt.title("Sales Distribution by Category")
plt.ylabel("")
plt.show()
```

```python
df["Order_Date"] = pd.to_datetime(df["Order_Date"])
sales = df.groupby("Order_Date")["Total_Sales"].sum()

plt.plot(sales.index, sales.values)
plt.title("Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.show()
```

```python
df.boxplot(column="Total_Sales", by="Category")
plt.title("Total Sales by Category")
plt.suptitle("")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
plt.scatter(df["Unit_Price"], df["Total_Sales"])
plt.title("Unit Price vs Total Sales")
plt.xlabel("Unit Price")
plt.ylabel("Total Sales")
plt.show()
```

```python
df.groupby("Category")["Total_Sales"].sum().plot(kind="bar")
plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```
